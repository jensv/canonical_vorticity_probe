```python
import mst_mach_211028_v2 as mst
import matplotlib.pyplot as plt
import matplotlib.offsetbox as mo
import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("paper")
```

```python
result = mst.mach_results(1210818038, nsmooth=100, tmin=0.015, tmax=0.045)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(5, 4))
axes[0].plot(result["t"]*1e3, result["r"], label="r")
axes[0].set_ylabel("v [M]")
axes[0].axes.xaxis.set_ticklabels([])
#axes[0].axes.add_artist(mo.AnchoredText("r", loc="lower right", frameon=False))
axes[0].set_title(r"$r$")
axes[1].plot(result["t"]*1e3, result["theta"], label=r"$\theta$")
axes[1].set_ylabel("v [M]")
axes[1].axes.xaxis.set_ticklabels([])
#axes[1].axes.add_artist(mo.AnchoredText(r"$\theta$", loc="upper right"))
axes[1].set_title(r"$\theta$")

axes[2].plot(result["t"]*1e3, result["phi"], label=r"$\phi$")
axes[2].set_ylabel("v [M]")
axes[2].set_xlabel("t [ms]")
axes[2].set_title(r"$\phi$")
#axes[2].set_ylim(-1.5, 0.9)
#axes[2].axes.add_artist(mo.AnchoredText(r"$\phi$", loc="lower right", frameon=False))
plt.tight_layout()
plt.savefig("../../output/1210818038.eps", dpi=600)
```
